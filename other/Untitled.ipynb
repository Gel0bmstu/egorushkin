```python
import matplotlib.pyplot as plt
import random
import time

def plot(data, xlabel, ylabel, title, legend):
    width = 20
    height = 12
    
    plt.figure(figsize=(width, height))    
    for each in data:
        plt.plot(each)
    plt.suptitle(title, fontsize=30)
    plt.xlabel(xlabel, fontsize = 30)
    plt.ylabel(ylabel, fontsize = 30)
    plt.xticks(fontsize = 20)
    plt.yticks(fontsize = 20)
    plt.legend(legend, prop={'size': 20})
    plt.grid()
    plt.show()
    
    
class dihonestCoin:
    heads_count = 0
    dishonest_heads_count = 0
    result = []
    dishonest_result = []
    
    flips_count = 0
    
    def flip(self):
        dishonest_value = random.randrange(1, 100, 1)
        value = random.randrange(0, 2, 1)
        
        if dishonest_value < 55:
            self.dishonest_heads_count = self.dishonest_heads_count + 1
        
        if value == 0:
            self.heads_count = self.heads_count + 1
        
        self.flips_count = self.flips_count + 1
        
        self.dishonest_result.append(self.dishonest_heads_count/self.flips_count)
        self.result.append(self.heads_count/self.flips_count)
    
    def plot_result(self):
        plot([self.result, self.dishonest_result], 'probability', \
             'iterations', 'flips', ['flips', 'dishonest flips'])
    
    def experiment(self, iterations):
        for i in range(iterations):
            self.flip()
        
        self.plot_result()

n = 6000
        
dc = dihonestCoin()
dc.experiment(n)
print(dc.result[n-1])
print(dc.dishonest_result[n-1])
```
